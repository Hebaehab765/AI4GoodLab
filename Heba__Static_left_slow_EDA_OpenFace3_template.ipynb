üîç 1. Plot Gaze Over Time
Goal: Visualize where the person is looking over time.

‚úÖ Why it's useful:
gaze_yaw and gaze_pitch indicate the horizontal and vertical angles of gaze (in radians), respectively.

Visualizing this over time helps identify:

- Attention shifts

- Distraction (rapid, large shifts)

Spikes: Sudden eye movement

Yaw > 0: Looking right, Yaw < 0: Looking left

Pitch > 0: Looking up, Pitch < 0: Looking down

üßç‚Äç‚ôÇÔ∏è 3. Parse Face Detection & Landmarks
Goal: Extract detailed face location or landmark data for spatial analysis.

‚úÖ Why it's useful:
The face_detection field gives you bounding box and confidence.

The landmarks field includes precise facial keypoints.

You can use them to:

Calculate face size/distance (proxy for head movement)

Track specific landmark changes (e.g., mouth, eyes)

üîß How to do it:
First, parse the stringified lists:

```python
print(df['face_detection'].iloc[0])
```

```python
import ast
df['face_detection_parsed'] = df['face_detection'].apply(ast.literal_eval)
```

Then extract useful values:


```python
df['face_x'] = df['face_detection_parsed'].apply(lambda x: x[0])  # x-coordinate of face box
df['face_confidence'] = df['face_detection_parsed'].apply(lambda x: x[4])  # detection confidence
```

```python
# Example: x, y, x2, y2, confidence (first 5 values)
df['x'] = df['face_detection_parsed'].apply(lambda x: x[0])
df['y'] = df['face_detection_parsed'].apply(lambda x: x[1])
df['x2'] = df['face_detection_parsed'].apply(lambda x: x[2])
df['y2'] = df['face_detection_parsed'].apply(lambda x: x[3])
df['confidence'] = df['face_detection_parsed'].apply(lambda x: x[4])
```

üí™ 4. Parse Action Units (AUs)
Goal: Analyze facial muscle movements linked to specific expressions.

‚úÖ Why it's useful:
AUs come from the Facial Action Coding System (FACS).

Each AU corresponds to a muscle activation:

e.g., AU01 = Inner brow raise, AU12 = Smile

Intensities reflect how strongly those muscles were activated.

üîß How to do it:

```python
df['action_units_parsed'] = df['action_units'].apply(ast.literal_eval)
df['AU01_intensity'] = df['action_units_parsed'].apply(lambda x: x[0][0])  # example AU index
```

```python
import ast

df['au_values'] = df['action_units'].apply(ast.literal_eval)

df['au_values'] = df['au_values'].apply(lambda x: x[0] if isinstance(x, list) and len(x) == 1 else x)

au_df = df['au_values'].apply(pd.Series)
au_df.columns = [f"AU_{i}" for i in range(au_df.shape[1])]

# Join back to original DataFrame
df = pd.concat([df, au_df], axis=1)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
for i in range(3):  # Show AU_0, AU_1, AU_2 as examples
    plt.plot(df['datetime'], df[f"AU_{i}"], label=f"AU_{i}")
plt.legend()
plt.title("Sample Action Unit Activations Over Time")
plt.xlabel("Time")
plt.ylabel("AU Intensity")
plt.grid(True)
plt.show()
```

üß† Interpretation
Each AU_x represents a learned facial muscle feature. Spikes might correspond to:

AU_0, AU_1: Brow raises, frowns

AU_5‚ÄìAU_7: Eye squints, cheek raises

AU_6+: Lip tightening, jaw drop

Though OpenFace doesn't label each AU directly in this TSV, these features can be clustered or correlated with emotion or task performance over time.

üß† Goal
Use features like gaze, head pose, facial action units (AUs), and eye behavior extracted by OpenFace to infer indicators of focus or concentration.
### üîç Key Behavioral Indicators of Focus (based on literature and HCI studies)

| **Feature Category** | **Focus Indicators (High)**                              | **Distraction Indicators (Low)**                      |
|----------------------|----------------------------------------------------------|-------------------------------------------------------|
| **Gaze Stability**   | Low variance in yaw/pitch; gaze toward screen            | High gaze shifts, frequent yaw/pitch changes          |
| **Head Pose**        | Low motion, consistent forward pose                      | Frequent tilting, looking away                        |
| **Eye AUs**          | AU45 (blink), AU1/2 (brow raise), AU4 (frown)            | Excessive blinking (AU45), relaxed face (low AUs)     |
| **Mouth AUs**        | Tension: AU10 (upper lip), AU14 (dimpler)                | Talking/yawning: AU25/26/27                           |
| **Emotion (optional)** | Neutral or mild tension                                | Surprised, sad, or bored                              |


 1. Gaze Behavior

```python
df[['gaze_yaw', 'gaze_pitch']].describe()
df[['gaze_yaw', 'gaze_pitch']].plot()
```

Interpretation: Low variance = likely focused. High variance or spikes = distraction or screen-switching.



2. Blink Rate (AU45)

```python
df['AU_blink'] = df['au_values'].apply(lambda x: x[6])  # Adjust index as needed for AU45
df['AU_blink'].plot()
```

Interpretation: Blinks are natural, but excessive blinking or clusters can signal fatigue or cognitive overload.

Use rolling().mean() to compute blink trends.

3. Tension/Attention AUs

Look at:

AU1, AU2 (brow raise)

AU4 (brow furrow = effort/focus)

AU7 (lid tightener)

AU14 (dimpler = tension)

```python
focus_aus = df[[f"AU_{i}" for i in [0, 1, 2, 5, 6, 7]]]
focus_aus.plot(subplots=True, figsize=(10, 8), title='Attention-Related AUs Over Time')
```

High AU4 + AU7 + stable gaze often implies mental effort or concentration.



**4. Detect Focus Periods (simple rule-based)**

```python
# Example rule: focused if gaze steady + AU4 > threshold
df['is_focused'] = (
    (df['gaze_yaw'].abs() < 0.1) &
    (df['gaze_pitch'].abs() < 0.1) &
    (df['AU_2'] > 0.01) &   # Brow raise
    (df['AU_4'] > 0.01)     # Brow lower (concentration)
)

df['is_focused'].astype(int).plot()
```

This binary label can be refined and used for ML training later.

5. Create a Summary

```python
focus_ratio = df['is_focused'].sum() / len(df)
print(f"Estimated focus ratio: {focus_ratio:.2%}")
```

üß† Final Tips for Interpretation
Sustained attention is reflected by: stable gaze + steady facial tension (e.g., AU4, AU7).

Cognitive overload or frustration might show as: gaze aversion + AU9 (nose wrinkle) or AU23/24 (lip tightening).

Combine temporal smoothing (rolling windows) to reduce false shifts.
